```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Data pengukuran
strain = np.array([0.00, 0.05, 0.10, 0.15, 0.20, 0.25, 0.30, 0.35, 0.40, 0.45, 0.50])
stress = np.array([0.000, 0.252, 0.531, 0.840, 1.184, 1.558, 1.975, 2.444, 2.943, 3.500, 4.115])

# Regresi linear
slope, intercept, r_value, p_value, std_err = linregress(strain, stress)

# Persamaan regresi linear:  = a + b
a = intercept
b = slope

# Koefisien korelasi ()
rho = r_value

# Plotting data pengukuran
plt.scatter(strain, stress, label='Pengukuran')

# Plotting regresi linear
regression_line = a + b * strain
plt.plot(strain, regression_line, label=f'Regresi Linear\n = {a:.3f} + {b:.3f}\n = {rho:.3f}', color='red')

# Menambahkan label dan judul
plt.xlabel('Strain')
plt.ylabel('Stress')
plt.title('Stress vs Strain with Linear Regression')

# Menampilkan legenda
plt.legend()

# Menampilkan plot
plt.show()

# Plotting hanya data pengukuran
plt.scatter(strain, stress, label='Pengukuran')
plt.xlabel('Strain')
plt.ylabel('Stress')
plt.title('Stress vs Strain - Pengukuran')

# Menampilkan legenda
plt.legend()

# Menampilkan plot
plt.show()

# Menampilkan nilai a, b, dan 
print(f'Nilai a: {a:.3f}')
print(f'Nilai b: {b:.3f}')
print(f'Koefisien Korelasi (): {rho:.3f}')
```
