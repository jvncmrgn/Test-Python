```python
import matplotlib.pyplot as plt
import numpy as np

# Fungsi g(omega)
def g(omega):
    return delta_N / delta_omega

# Definisikan nilai delta_N dan delta_omega sesuai kebutuhan Anda
delta_N = 5
delta_omega = np.linspace(0.1, 10, 100)  # Contoh: 100 titik dari 0.1 hingga 10

# Hitung g(omega)
hasil_g = g(delta_omega)

# Buat plot
plt.plot(delta_omega, hasil_g)
plt.xlabel('Delta Omega')
plt.ylabel('g(omega)')
plt.title('Grafik dari g(omega) = delta N / delta omega')
plt.grid(True)
plt.show()
```
