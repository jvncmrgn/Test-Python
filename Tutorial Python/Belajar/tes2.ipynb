```python
import matplotlib.pyplot as plt
import numpy as np

# Definisikan nilai v dan k sesuai kebutuhan Anda
v = np.linspace(0.1, 10, 100)  # Contoh: 100 titik dari 0.1 hingga 10
k = 2  # Contoh: Nilai k

# Hitung nilai omega
omega = v * k

# Buat plot
plt.plot(v, omega)
plt.xlabel('v')
plt.ylabel('omega')
plt.title('Grafik dari omega = v * k')
plt.grid(True)
plt.show()
```
